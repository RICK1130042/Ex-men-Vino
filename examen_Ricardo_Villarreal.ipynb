# 1. Estamos interesados sólo en los vinos de Italia, España, México y Chile. Escribe el código necesario para obtener un DataFrame que contenga sólo las filas de interés.

```python
# 1
import pandas as pd
vino = pd.read_csv('wine.csv')

# Filtrar solo los vinos de Italia, España, México y Chile
paises_interes = ["Italy", "Spain", "Mexico", "Chile"]
df_filtrado = vino[vino["country"].isin(paises_interes)]

# Nuevo data frame
df_filtrado.to_csv('Chile-Italy-Spain-Mexico.csv', index=False)


ejercicio1 = pd.read_csv('Chile-Italy-Spain-Mexico.csv')
ejercicio1
```

# 2. Escribe el código necesario para encontrar de qué país se tienen más reseñas de vinos y cuántas son.

```python
#2

# Contar el número de reseñas por país
reseñas_por_pais = df_filtrado["country"].value_counts()

# Obtener el país con más reseñas y la cantidad
pais_mas_reseñas = reseñas_por_pais.idxmax()
cantidad_mas_reseñas = reseñas_por_pais.max()

# Mostrar el resultado
print(f"El país con más reseñas es {pais_mas_reseñas} con {cantidad_mas_reseñas} reseñas.")

display(reseñas_por_pais)
```

# 3. Escribe el código necesario para generar una tabla en dónde podamos ver la mediana de las calificaciones de los vinos de cada región indicando también la provincia.

```python
#3

# Calcular la mediana de las calificaciones agrupando por país y provincia
mediana_calificaciones = df_filtrado.groupby(["country", "province"])["points"].median().reset_index()

display(mediana_calificaciones)
```

# 4. Genera una tabla similar al punto anterior, pero ahora muestra el promedio y desviación estándar en una sola tabla.

```python
#4

# Calcular el promedio y la desviación estándar de las calificaciones agrupando por país y provincia
estadisticas_calificaciones = df_filtrado.groupby(["country", "province"])["points"].agg(['mean','std']).reset_index()


display(estadisticas_calificaciones)
```

# 5. Considera que un buen vino tiene una reseña con una puntuación por lo menos de 80. De estos vinos, ¿cuál tiene la mejor relación precio-calidad?

```python
#5

# Filtrar vinos con puntuación de al menos 80
vinos_buenos = df_filtrado[df_filtrado["points"] >= 80].copy()

# Calcular la relación precio-calidad (Calificación / Precio)
vinos_buenos["Relación Precio-Calidad"] = vinos_buenos["points"] / vinos_buenos["price"]

# Obtener el vino con la mejor relación precio-calidad
mejor_vino = vinos_buenos.loc[vinos_buenos["Relación Precio-Calidad"].idxmax()]

mejor_vino_df = pd.DataFrame([mejor_vino])

display(mejor_vino_df)
```

# 6. ¿Qué tan confiable es la reseña del catador del punto anterior? Considera que el catador es confiable si su promedio de calificaciones tiene una diferencia de menos de una desviación estándar del promedio de reseñas para los vinos del país del que procede el vino del punto anterior.   

```python
#6

# Obtener el país y el catador del mejor vino
pais_mejor_vino = mejor_vino["country"]
catador_mejor_vino = mejor_vino["taster_name"]

# Calcular el promedio y desviación estándar de calificaciones para los vinos del mismo país
estadisticas_pais = df_filtrado[df_filtrado["country"] == pais_mejor_vino]["points"].agg(['mean', 'std'])
promedio_pais = estadisticas_pais['mean']
desviacion_pais = estadisticas_pais['std']

# Calcular el promedio de calificaciones del catador
promedio_catador = df_filtrado[df_filtrado["taster_name"] == catador_mejor_vino]["points"].mean()

# Evaluar confiabilidad
diferencia = abs(promedio_catador - promedio_pais)
es_confiable = diferencia < desviacion_pais

# Mostrar los resultados
resultado_confiabilidad = {
    "Catador": catador_mejor_vino,
    "País del Vino": pais_mejor_vino,
    "Promedio del País": promedio_pais,
    "Desviación del País": desviacion_pais,
    "Promedio del Catador": promedio_catador,
    "Diferencia": diferencia,
    "Es Confiable": es_confiable
}


# Convertir en DataFrame para visualizarlo mejor
resultado_df = pd.DataFrame([resultado_confiabilidad])

display(resultado_df)
```
